```python
import networkx as nx
import matplotlib.pyplot as plt


# 2部グラフの作成
G = nx.Graph()

# ユーザと商品のノードを追加
users = df_log['User'].unique()
items = df_log['Item'].unique()

G.add_nodes_from(users, bipartite=0, color='lightblue')  # ユーザは0の属性を持ち、色はlightblue
G.add_nodes_from(items, bipartite=1, color='lightgreen')  # 商品は1の属性を持ち、色はlightgreen

# エッジを追加
edges = [(row['User'], row['Item']) for _, row in df_log.iterrows()]
G.add_edges_from(edges)

# ノードの色を取得
node_colors = [node[1]['color'] for node in G.nodes(data=True)]

# 2部グラフの可視化
user_nodes = set(users)
pos = {node: (0, int(node) - 1) if node in user_nodes else (1, ord(node) - 65) for node in G.nodes()} 

plt.figure(figsize=(6, 6))
nx.draw(G, pos, with_labels=True, font_weight='bold', node_color=node_colors, node_size=700)
plt.title('Bipartite Graph of Users and Items')
plt.show()
```

```python
# プロジェクション
G_user = nx.bipartite.projected_graph(G, users)
G_item = nx.bipartite.projected_graph(G, items)

# ノードの色を取得
user_colors = [node[1]['color'] for node in G_user.nodes(data=True)]
item_colors = [node[1]['color'] for node in G_item.nodes(data=True)]

# プロジェクション後のネットワークの可視化
plt.figure(figsize=(6, 4))

# ユーザネットワークの可視化
plt.subplot(1, 2, 1)
nx.draw(G_user, with_labels=True, font_weight='bold', node_color=user_colors, node_size=700)
plt.title('User Network Projection')

# 商品ネットワークの可視化
plt.subplot(1, 2, 2)
nx.draw(G_item, with_labels=True, font_weight='bold', node_color=item_colors, node_size=700)
plt.title('Item Network Projection')

plt.tight_layout()
plt.show()
```
